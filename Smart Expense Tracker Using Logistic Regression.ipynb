**Data** **Cleaning**

```python
def clean_text(text):

    cleaned_text = re.sub(r'[^A-Za-z0-9\s]','', text)  # Remove special characters
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()  # Remove extra spaces
    cleaned_text = re.sub(r'\d+', '', cleaned_text)  # Remove digits

    return cleaned_text

data['Cleaned_Sentence'] = data['Sentence'].apply(clean_text)
```

```python
data.head()
```

```python
# Unicode Normalization
def normalize_unicode(text):
    return re.sub(r'[^\x00-\x7F]+', '', text)  # Remove non-ASCII characters

data['Normalized_Text'] = data['Cleaned_Sentence'].apply(normalize_unicode)
```

```python
# Step 9: Stop-word Removal
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# Download the stopwords resource
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

data['Stopword_Removed_Text'] = data['Word_Tokens'].apply(remove_stopwords)
```
